```python
#Check the count for each category in the "Registration State" column
df["Registration State"].value_counts()
```

```python
#Violation county and their count
df["Violation County"].value_counts().plot(kind="bar")
plt.title("Value counts of Violation County")
plt.xlabel("Violation County")
plt.xticks(rotation=0)
plt.ylabel("Count")
plt.show()
```

```python
#A Scatter Plot to show the relationship between Violation Time and Violation County 
sns.scatterplot(x="Violation Time", y="Violation County", data=df, palette="Dark2", s=80)
plt.title("Relationship between Violation Time and Violation County ")
plt.show()
```
